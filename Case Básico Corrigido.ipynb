<br/> **Dado um número natural n qualquer, escreva todos os numeros inteiros de 1 até n. Para todos os números para os quais n não seja divísível por, pinte-os de vermelho. Caso n seja divisível, pinte-os de verde. Por fim, diga se n é um número primo ou não, contando quantas vezes o número foi divisível.**

Ajude Alice! Vocês devem:
- Escrever um programa que leia um número natural 'n' e diga se ele é ou não um número primo
- Printe todos os números de 1 a 'n' no processo (em sequência na mesma linha)
- Se n não for divisível pelo número, destaque com fundo vermelho, caso contrário destaque com a cor verde. Exemplo de output esperado:



```python
#Retirando informação do usuário
núm = int(input('Digite um número: '))
#No início, a quantidade de vezes que o número foi dividido deve ser 0
total = 0
#O código deverá ser executado a partir do número 1 até o número inserido anteriormente. O +1 deve-se ao fato de que o python sempre conta um a menos
for c in range (1, núm + 1):
#O código aqui deverá retornar, em verde,os números pelos quais o resto da divisão do número inserido pelo usuário foi igual a 0
    if núm % c == 0:
        print('\033[32m',end='')
#Deve ser somado 1 ao total cada vez que o número inserido foi dividido por outro
        total +=1
#O código aqui deverá retornar, em vermelho,os números pelos quais o resto da divisão do número inserido pelo usuário foi diferente de 0
    else:
        print ('\033[31m',end='')
#Escrever a contagem dos números 
    print('{} '.format(c), end='')
#Agora o código deverá retornar a resposta final, ou seja, a quantidade de vezes que o número foi dividido. Caso tenha sido somente duas vezes, também deve retornar que o número em questão é primo, tendo em vista que números primos só são divisíveis por 1 e por eles mesmos.   
print('\n\033[mO número {} foi divisível {} vezes'.format(núm,total))
if total==2:    
    print('SENDO um número primo')
else:
    print ('NÃO sendo um número primo')
```

## **Ex. 11 - Progressão Infinita** 
*Cansada de ser bombardeada por desafios pelo Chapeleiro Maluco sem conseguir resolvê-los, Alice resolveu estudar matemática. A Lagarta Azul recomendou um livro, escrito pelo Chapeleiro Maluco, feito para que as pessoas treinassem um pouco para seus desafios. Usando o livro, Alice chegou ao assunto "progressão aritmética", e gostaria da sua ajuda. Ela não consegue, dado o primeiro termo e a razão de uma PA, construir a progressão. Assim, ela quer sua ajuda para conseguir resolver alguns exercícios nos quais precisa construir progressões aritméticas com 10 ou mais termos.*

Sua tarefa: <br />

- Desenvolva um programa que leia o primeiro termo e a razão de uma PA. Após isso, **printe os 10 primeiros termos dessa progressão.**
- Pergunte para o usuário: **Quantos termos a mais você quer printar?**
  - Se o input for um número maior que 0: **printe essa quantidade a mais de termos de acordo com o último mostrado**
  - Se o input for 0, o programa encerrará
- Ao final mostre quantos termos foram printados: **Progressão finalizada com xx termos printados**

### Exemplo de output esperado: 



```python
#Retirando informações do usuário
primeiro = int(input('Primeiro termo: '))
razão = int(input('Razão: '))
#Para realizarmos a contagem precisaremos das variável termo, cont para sabermos quantos termos já foram retornados e total, para saber quantos termos devem ser retornados. Para que no início o total de termos seja igual a 10, usaremos a variável "mais" com esse valor
termo = primeiro
total = 0
cont = 1
mais = 10
#Enquanto o usuário não colocar "0", ou seja, não deseja ver mais termos da PA, o programa deve continuar
while mais != 0:
#O programa deve retornar a quantidade desejada pelo usuário, para isso utilizaremos as variáveis total e mais
    total = total + mais
    #Enquanto a contagem não chegar a 10 termos o programa continuará sendo executado
    while cont <= total:
        print('{} -> '. format(termo), end='')
    #Cada vez que o programa for executado deverá somar a razão no próximo termo
        termo += razão
        cont += 1
    print ('PAUSA')
    #Perguntar ao usuário se ele deseja mostrar mais termos
    mais = int(input('Quantos termos você quer mostrar a mais? '))
print ('FIM')

            
```

*O cartão de crédito é um instrumento super sofisticado. Apesar de parecer que os números são puramente aleatórios, eles seguem uma lógica que automaticamente o invalida caso algum número seja digitado errado. Muito mais fácil torrar o dinheiro no Shopee assim né.*

Nesse exercício, você irá simular esse algoritmo de validação. Para isso, lembre-se que:

* 1. Um cartão é composto de uma sequência de 13 a 16 dígitos decimais

* 2. Segundo a determinação, cartões de crédito e débito só podem começar com os dígitos 3, 4, 5 ou 6

* 3. Cálculo de validação do número:
    - 3.1 Iniciando com o dígito mais à direita, multiplique por 2 (dois) todos os dígitos de posição par, e deixe inalterados os dígitos das posições ímpares. 
    
      - Por exemplo: o número 49927398716 seria reescrito como 4(18)9(4)7(6)9(16)7(2)6. Os parênteses ilustram os dígitos que foram dobrados, e são usados apenas para fins didáticos. Sem os parênteses, o número resultante seria 4189476916726.
    - 3.2 Some todos os dígitos, independente da sua posição
      - Assim, o número resultante do passo 1 produziria: 4+1+8+9+4+7+6+9+1+6+7+2+6 = 70
    - 3.3 Se a divisão por dez produzir um resto igual a zero, então o número inicial é válido. Note que, obviamente, isso não quer dizer que o cartão seja legítimo, ou que tenha crédito, ou que não tenha expirado. Passar no teste quer simplesmente dizer que não ocorreram erros, ou seja, o número é bem formado.

- 4. Os dígitos mais significativos (prefixos) identificam qual a companhia que opera o cartão, de acordo com o dicionário criado abaixo



- **De acordo com os parâmetros acima faça uma função chamada credit() que valide o número de cartão de crédito e retorne**
    - Se o número é válido
        - Mostrar a quantidade de dígitos - Deve no range correto (13 ~ 16)
        - Mostrar qual o primeiro dígito - Deve ser 3, 4, 5 ou 6
        - Mostrar se possui apenas números - Deve conter apenas números e recusar caracteres especiais e/ou strings em geral
    - Qual a bandeira/ Operadora do cartão se possuir, caso contrário a Operadora é "Desconhecida"


```python
# Dicionário com informações necessárias
operadora = [{'Operadora':'Mastercard',  'Prefixo': [51, 52, 53, 54, 55], 'Tamanho': 16},
             {'Operadora':'Visa',        'Prefixo': [4]                 , 'Tamanho': [13, 16]},
             {'Operadora':'Amex',        'Prefixo': [34, 37]            , 'Tamanho': 16}, 
             {'Operadora':'Diners',      'Prefixo': [30, 36, 38]        , 'Tamanho': 16},
             {'Operadora':'Discover',    'Prefixo': [6011]              , 'Tamanho': 16},
             {'Operadora':'enRoute',     'Prefixo': [2014, 2149]        , 'Tamanho': 16},
             {'Operadora':'JCB',         'Prefixo': [3]                 , 'Tamanho': 16},
             {'Operadora':'Dev Express', 'Prefixo': [6666]              , 'Tamanho': 16}]

def credit(n=str):
    n = str(n)
    # 1. Verificando se o número é válido
    # 1.1 Quantidade de dígitos
    qt_digits = len(n)
    qt_digit_status = False
    if 12 < qt_digits < 17: qt_digit_status = True # Verifica range da qt. de dígitos

    # 1.2 Primeiro dígito
    first_digit = n[0]
    first_digit_status = False
    if first_digit in '3456': first_digit_status = True # Verifica primeiro dígito
        
    # 1.3 Conter apenas números
    only_numbers = False
    if n.isnumeric(): only_numbers = True 

    # 1.4 Cálculo da validade do número
    # 1.4.1 Dobrando os números nas posições desejadas
    n_validation_status = False # Inicializando a variável final de status sendo falsa por padrão
    
    if qt_digit_status & only_numbers == True : # Só faz a conta se houverem apenas dígitos numéricos
        new_n = '' # Variável que armazena o número tratado final
        #Se a quantidade de dígitos for par (daria para fazer no mesmo loop, mas quis separar):
        if qt_digits % 2 == 0:
            for pos, digit in enumerate(n):
                if pos % 2 == 0:
                    new_n += str(int(digit) * 2)
                else:
                    new_n += digit

        # Se a quantidade de dígitos for ímpar:
        elif qt_digits % 2 ==1:
            for pos, digit in enumerate(n):
                if pos % 2 == 0:
                    new_n += digit
                else:
                    new_n += str(int(digit) * 2)

        # 1.4.2 Somando todos os digitos individualmente
        sum_n = 0
        for digit in new_n:
            sum_n += int(digit)

        # 1.4.3 Se a soma de todos os digitos / 10 tiver um resto 0, os números são válidos
        if sum_n % 10 == 0: n_validation_status = True
    
    # 1.5 Conclusão sobre a validade do cartão (Prepare os olhos, quem não tá acostumado com esses códigos fica um pouco intimidador)
    
    # A n_validation_status já cobre as outras condições, mas não é necessário o primeiro dígito ser válido para fazer a conta
    if  first_digit_status & n_validation_status == True: 
        print(f'\n\033[1mCARTÃO VÁLIDO - Resumo dos parâmetros pedidos:\033[0m\n')    
        print(f'\033[7;32m1. O código {n} é bem formado, após dobrar os caracteres indicados temos {new_n}, a soma de todos indivualmente é {sum_n} logo a divisão por 10 gera o resto zero\033[0m')
    else:
        print(f'\n\033[1mCARTÃO INVÁLIDO - Resumo dos parâmetros pedidos:\033[0m\n')
        print(f'\033[31m1. O código {n} não é bem formado e/ ou as condições precedessoras não são válidas\033[0m')

    if qt_digit_status: print(f'\033[7;32m2. Quantidade de dígitos: {qt_digits}\033[0m')
    else: print(f'\033[31m2. Quantidade de dígitos: {qt_digits}\033[0m')
        
    if first_digit_status: print(f'\033[7;32m3. Primeiro dígito: {first_digit}\033[0m')
    else: print(f'\033[31m3. Primeiro dígito: {first_digit}\033[0m')
    
    if only_numbers == True: print(f'\033[7;32m4. Todos os caracteres numéricos: {only_numbers}\033[0m')
    else: print(f'\033[31m4. Todos os caracteres numéricos: {only_numbers}\033[0m')
    
    
    # 2. Achando a operadora - O cartão pode não ser totalmente válido, mas ainda sim dá pra achar a operadora
    prefix = 0 # Inicializando variável
    # 2.1 Achando prefixo - dependendo do primeiro dígito existe um padrão da qt. de dígitos no prefixo
    if first_digit == '6':
        prefix = n[:4]
    elif first_digit == '3' or first_digit == '4':
        prefix = first_digit
    elif first_digit == '5':
        prefix = n[:2]
    else:
        prefix = 0 # Condição a qual se torna inválido
    #print(f'prefixo: {prefix}') # Teste
    
    # 2.2 Percorrendo dicionário para de fato encontrar a operadora
    card_operator = 'Desconhecida'
    for item in operadora:
        # Se encontrar o prefixo armazenado dentro do dicionário, retorna a operadora correspondente
        if int(prefix) in item['Prefixo']:
            card_operator = item['Operadora']
    # Resposta final Operadora - Coloquei separado dos outros pela construção da solução, mas não poderia ser junto
    if card_operator == 'Desconhecida': print(f'\033[31m5. Operadora: {card_operator}\033[0m')
    else: print(f'\033[7;32m5. Operadora: {card_operator}\033[0m')
    

```

- Caso não for usar o alfabeto:

#### Para testar se deu certo comece colocando uma mensagem fácil como 'abcdefg...xyz' e passe o shift=1, se as letras pularem uma posição então sucesso 😎. Últimos pontos para facilitar sua vida:

- O código deve funcionar com letras maiúsculas e minúsculas
- Não é preciso codificar caracteres especiais como *&#@!, ignore-os
- Logo, tome cuidado com o espaço entre as letras, também devem ser ignorados

```python
# Mode = 0 Significa que quero codificar uma mensagem
# Mode = 1 o contrário
def caesar_cipher(msg, shift=0, mode=0):
    result = ""
 
    # Iterando sobre as palavras
    for letra in msg:
        # Checando se é alfanumérico
        if letra.isalpha():
              # Para letras maiúsculas
            if letra.isupper():
                if mode == 0:
                    result += chr((ord(letra) + shift - 65) % 26 + 65)
                else:
                    result += chr((ord(letra) - shift - 65) % 26 + 65)
             # Para letras minúsculas
            elif letra.islower():
                if mode == 0:
                    result += chr((ord(letra) + shift - 97) % 26 + 97)
                else:
                    result += chr((ord(letra) - shift - 97) % 26 + 97)
                    
        # Se não afanumérico apenas adicionar a letra na mensagem final
        else:
            result += letra
            
    return result
# Done
a = 'abcdefghijklmnopqrstuvwxyz'
b = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# Testando
encoded = caesar_cipher(a, 1, mode=0)
print(encoded)
encoded = caesar_cipher(b, 1, mode=0)
print(encoded)





```
